```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('Ecoli_v2.csv')
sns.countplot(x="name", data=df)
```

```python
df = df.replace({'site1':1,'site2':2,'site3':3,'site4':4})
X = df.iloc[:,:6]
```

```python
kmeans = KMeans(n_clusters=4,random_state=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
predict_k4 = kmeans.inertia_
df['predict_k4'] = y_kmeans+1
df['predict_k4'] = df['predict_k4'].replace({2:3,3:2})
```

```python
cm = confusion_matrix(df['name'].values,df['predict_k4'].values)
plt.figure(figsize=(8,6),dpi=80)
sns.heatmap(cm,cmap='rocket_r',xticklabels=[1,2,3,4],yticklabels=[1,2,3,4],annot=True)
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
def plot(df,eps,min_samples):
    fig = plt.figure(figsize=(5,5))
    X = df[['x','y']]
    db = DBSCAN(eps, min_samples)
    df['predict_db'] = db.fit_predict(X) +2
    for p in range(1,len(list(set(df['predict_db'])))+1):
        plt.scatter(df[df['predict_db']==p]['x'], df[df['predict_db']==p]['y'], label=str(p))
        plt.legend()
    plt.show()
    
    
    cm = confusion_matrix(df['class'].values,df['predict_db'].values)
    plt.figure(figsize=(5,5),dpi=80)
    sns.heatmap(cm,cmap='rocket_r',xticklabels=[1,2,3,4],yticklabels=[1,2,3,4],annot=True)
    
    from sklearn.metrics.cluster import homogeneity_score
    print(round(homogeneity_score(df['class'].values,df['predict_db'].values),3))
```
